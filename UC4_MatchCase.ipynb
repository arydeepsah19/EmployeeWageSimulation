# UC4 – Daily Wage Using match-case

```python
import random

# Constants
WAGE_PER_HOUR = 20
FULL_DAY_HOURS = 8
PART_TIME_HOURS = 4

emp_check = random.randint(0, 2)

match emp_check:
    case 1:
        hours = FULL_DAY_HOURS
    case 2:
        hours = PART_TIME_HOURS
    case _:
        hours = 0

daily_wage = hours * WAGE_PER_HOUR
print("Daily Wage:", daily_wage)
```
