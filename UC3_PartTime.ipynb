# UC3 – Part-Time Employee Wage

```python
import random

# Constants
WAGE_PER_HOUR = 20
FULL_DAY_HOURS = 8
PART_TIME_HOURS = 4
IS_PRESENT = 1
IS_PART_TIME = 2

# Random attendance/type
emp_check = random.randint(0, 2)

if emp_check == IS_PRESENT:
    hours = FULL_DAY_HOURS
elif emp_check == IS_PART_TIME:
    hours = PART_TIME_HOURS
else:
    hours = 0

daily_wage = hours * WAGE_PER_HOUR
print("Daily Wage:", daily_wage)
```
