# UC2 - Daily Wage for Full-Time Employee

```python
import random

# Constants
WAGE_PER_HOUR = 20
FULL_DAY_HOURS = 8
IS_PRESENT = 1

emp_check = random.randint(0, 1)

if emp_check == IS_PRESENT:
    daily_wage = WAGE_PER_HOUR * FULL_DAY_HOURS
else:
    daily_wage = 0

print("Daily Wage:", daily_wage)
```
