# UC5 – Monthly Wage Calculation

```python
import random

# Constants
WAGE_PER_HOUR = 20
FULL_DAY_HOURS = 8
PART_TIME_HOURS = 4
WORKING_DAYS = 20

total_wage = 0

for day in range(WORKING_DAYS):
    emp_check = random.randint(0, 2)

    match emp_check:
        case 1:
            hours = FULL_DAY_HOURS
        case 2:
            hours = PART_TIME_HOURS
        case _:
            hours = 0

    total_wage += hours * WAGE_PER_HOUR

print("Monthly Wage:", total_wage)
```
