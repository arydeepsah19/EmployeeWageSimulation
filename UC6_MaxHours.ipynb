# UC6 – Wage with Maximum Hours and Maximum Days

```python
import random

# Constants
WAGE_PER_HOUR = 20
FULL_DAY_HOURS = 8
PART_TIME_HOURS = 4
MAX_HOURS = 100
MAX_DAYS = 20

total_hours = 0
total_days = 0
total_wage = 0

while total_hours < MAX_HOURS and total_days < MAX_DAYS:
    total_days += 1
    emp_check = random.randint(0, 2)

    match emp_check:
        case 1:
            hours = FULL_DAY_HOURS
        case 2:
            hours = PART_TIME_HOURS
        case _:
            hours = 0

    total_hours += hours
    total_wage += hours * WAGE_PER_HOUR

print("Total Wage:", total_wage)
```
